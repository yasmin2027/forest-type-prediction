```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix



df = pd.read_csv("Forest Types Dataset.csv")

print(df.isnull().sum())

x = df.drop('Cover_Type', axis=1)
y = df['Cover_Type']
y=y-1

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(x_train.shape)
print(x_test.shape)

#LOGISTICREGRESSION
log = LogisticRegression(max_iter=1000)
log.fit(x_train, y_train)
ylog_pred = log.predict(x_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, ylog_pred))
cm = confusion_matrix(y_test, ylog_pred)
cm_df = pd.DataFrame(cm, 
                     index=[f"Actual_{i}" for i in range(len(cm))],
                     columns=[f"Pred_{i}" for i in range(len(cm))])
print("Confusion Matrix:")
print(cm_df)


#DECISIONTREECLASSIFIER
dec = DecisionTreeClassifier(random_state=42)
dec.fit(x_train, y_train)
ydec_pred = dec.predict(x_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, ydec_pred))

cm = confusion_matrix(y_test, ydec_pred)
cm_df = pd.DataFrame(cm, 
                     index=[f"Actual_{i}" for i in range(len(cm))],
                     columns=[f"Pred_{i}" for i in range(len(cm))])
print("Confusion Matrix:")
print(cm_df)

#RANDOMFORESTCLASSIFIER
ran = RandomForestClassifier(random_state=42)
ran.fit(x_train, y_train)
yran_pred = ran.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, yran_pred))


cm = confusion_matrix(y_test, yran_pred)
cm_df = pd.DataFrame(cm, 
                     index=[f"Actual_{i}" for i in range(len(cm))],
                     columns=[f"Pred_{i}" for i in range(len(cm))])
print("Confusion Matrix:")
print(cm_df)


```
